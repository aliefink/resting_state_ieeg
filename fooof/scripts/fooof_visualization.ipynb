```python
def sort_elecs(periodic, anat, anat_guide):
    """
    Sort electrodes based on anatomical location and manual examination - only include WM, in-brain, ROI electrodes.

    Parameters
    ----------
    periodic : pandas.DataFrame
        Dataframe containing periodic data.
        @ Alie - this is the output of the fooof function. You want to adapt this to take the mne object instead and extract the ch_names. 
    anat : pandas.DataFrame
        Dataframe containing anatomical data.
    anat_guide : pandas.DataFrame
        Dataframe containing anatomical guide (Custom labels).

    Returns
    -------
    pandas.DataFrame
        Dataframe containing sorted electrodes.
    """ 

    # Convert 'label' column to lowercase
    anat['label'] = anat['label'].str.lower()

    # Merge MS001_periodic and MS001_anat
    periodic_anat = pd.merge(periodic, anat, left_on=periodic['ch_name'].str.split('-').str[0], right_on=anat['label'].str.split('-').str[0])

    # Replace 'unknown' values in 'YBA_1' column with values from 'ManualExamination'
    periodic_anat.loc[periodic_anat['YBA_1'].str.lower() == 'unknown', 'YBA_1'] = periodic_anat.loc[periodic_anat['YBA_1'].str.lower() == 'unknown', 'ManualExamination']

    # Filter out rows where 'ManualExamination' is 'wm', 'oob', 'white', or 'unknown'
    periodic_anat = periodic_anat[~periodic_anat['ManualExamination'].str.lower().isin(['wm', 'oob', 'white','unknown'])]

    # Make 'Long.name' column in anat_guide lowercase and strip whitespace
    anat_guide['Long.name'] = anat_guide['Long.name'].str.lower().str.strip()

    # Make 'YBA_1' column in periodic_anat lowercase and strip whitespace
    periodic_anat['YBA_1'] = periodic_anat['YBA_1'].str.lower().str.strip()

    # Merge MS001_periodic_anat with anat_guide
    periodic_anat = pd.merge(periodic_anat, anat_guide, left_on='YBA_1', right_on='Long.name')

    return periodic_anat

# Example usage:
processed_data = sort_elecs(MS001_periodic, MS001_anat, anat_guide)
len(processed_data[processed_data['Custom'] == 'OFC']['key_0'].unique())
```
